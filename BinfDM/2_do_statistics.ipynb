```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# binomial distribution
n = 10
p = 0.24
x = np.arange(n)

rv = stats.binom(n, p)
my_mean = rv.mean()
my_pmf = rv.pmf(x)

print(my_pmf)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
markerline, stemlines, baseline = ax.stem(x, my_pmf, markerfmt=" ")
plt.setp(baseline, color='k', linewidth=1)
plt.setp(stemlines, color='b', linewidth=1)
ax.vlines(my_mean, 0, max(my_pmf), color='red', linewidth=2)
ax.set_title("binomial PMF")
```

```python
# normal distribution
mu = 3
sigma = 0.1
x = np.arange(start=mu - 5*sigma, stop=mu + 5 * sigma, step=0.001)
rv = stats.norm(loc = mu, scale = sigma)

my_mean = rv.mean()
my_pdf = rv.pdf(x)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, my_pdf, color='b', linewidth=1)
ax.vlines(my_mean, 0, max(my_pdf), color='r', linewidth=2)
ax.set_title("Gaussian PDF")
```

```python
# =============================
# covariance and correlation
# =============================
x = np.random.normal(loc=1.0, scale=1.0, size=1000)
x_mean = np.mean(x)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(range(len(x)), x, color='blue', s=4)
ax.plot([0, len(x)], [x_mean, x_mean], color='red', linewidth=2)
ax.set_xlabel("index")
ax.set_ylabel('x')

y = np.random.normal(loc=2.0, scale=1.0, size=1000)
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(x, y, color='blue', s=4)
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
# calculate covariance using numpy
cov_matrix = np.cov(x, y)
print("covariance matrix between x and y = ")
print(cov_matrix)

# calculate covariance by definition
xy = x * y
xy.mean() - x.mean() * y.mean()

# OR
sum((x - x.mean()) * (y - y.mean())) / (len(x)-1) 

# correlation
cov_matrix[0, 1] / (x.std() * y.std())
print("correlation = ")
print(np.corrcoef(x, y))
```

```python
# determinant of correlation matrix 
x = np.random.normal(loc=1.0, scale=1.0, size=1000)
y = np.random.normal(loc=1.0, scale=1.0, size=1000)
z = np.random.normal(loc=1.0, scale=1.0, size=1000)

my_random_vector = np.stack((x, y, z))
print(my_random_vector.shape, "\n")
corrcoef_matrix = np.corrcoef(my_random_vector)
print(corrcoef_matrix.shape, "\n")
print(np.linalg.det(corrcoef_matrix), "\n")

my_random_vector = np.stack((x, y, z), axis=1)
print(np.corrcoef(my_random_vector))
print(my_random_vector.shape)

h = 2*x
my_random_vector = np.stack((x, y, h))
corrcoef_matrix = np.corrcoef(my_random_vector)
print(np.linalg.det(corrcoef_matrix))
```
